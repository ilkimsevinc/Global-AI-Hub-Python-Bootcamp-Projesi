```python
#Pizza sınıfı
class Pizza:
    def __init__(self):
        self.description = "Boş Pizza"
        self.cost = 0.0

    def get_description(self):
        return self.description

    def get_cost(self):
        return self.cost
```

```python
#Pizza altsınıfı
class KlasikPizza(Pizza):
    def __init__(self):
        self.description = "Klasik Pizza"
        self.cost = 25.0
        
    def get_description(self):
        return self.description

    def get_cost(self):
        return self.cost

class MargaritaPizza(Pizza):
    def __init__(self):
        self.description = "Margarita Pizza"
        self.cost = 30.00

    def get_description(self):
        return self.description

    def get_cost(self):
        return self.cost

class TurkishPizza(Pizza):
    def __init__(self):
        self.description = "Türk Pizza"
        self.cost = 35.0
        
    def get_description(self):
        return self.description

    def get_cost(self):
        return self.cost

class SadePizza(Pizza):
    def __init__(self):
        self.description = "Sade Pizza"
        self.cost = 30.0
        
    def get_description(self):
        return self.description

    def get_cost(self):
        return self.cost
```

```python
#Sos sınıfı
class Decorator(Pizza):
    def init(self, pizza):
        super().init()
        self.component = pizza

    def get_cost(self):
        return self.component.get_cost() + Pizza.get_cost(self)
    def get_description(self):
        return self.component.get_description() + ' ' + Pizza.get_description(self)
```

```python
#Sos alt sınıfları
class Zeytin(Decorator):
    def __init__(self, pizza):
        self.component = pizza
        self.price = 2.0
        self.description = "Zeytin"
    
    def get_cost(self):
        return self.component.get_cost() + self.price
    
    def get_description(self):
        return self.component.get_description() + ', ' + self.description
    
class Mantar(Decorator):
    def __init__(self, pizza):
        self.component = pizza
        self.price = 2.5
        self.description = "Mantar"
    
    def get_cost(self):
        return self.component.get_cost() + self.price
    
    def get_description(self):
        return self.component.get_description() + ', ' + self.description
    
class KeciPeyniri(Decorator):
    def __init__(self, pizza):
        self.component = pizza
        self.price = 3.0
        self.description = "Keçi Peyniri"
    
    def get_cost(self):
        return self.component.get_cost() + self.price
    
    def get_description(self):
        return self.component.get_description() + ', ' + self.description
    
class Et(Decorator):
    def __init__(self, pizza):
        self.component = pizza
        self.price = 4.0
        self.description = "Et"
    
    def get_cost(self):
        return self.component.get_cost() + self.price
    
    def get_description(self):
        return self.component.get_description() + ', ' + self.description
    
class Sogan(Decorator):
    def __init__(self, pizza):
        self.component = pizza
        self.price = 1.5
        self.description = "Soğan"
    
    def get_cost(self):
        return self.component.get_cost() + self.price
    
    def get_description(self):
        return self.component.get_description() + ', ' + self.description
    
class Misir(Decorator):
    def __init__(self, pizza):
        self.component = pizza
        self.price = 1.5
        self.description = "Mısır"
    
    def get_cost(self):
        return self.component.get_cost() + self.price
    
    def get_description(self):
        return self.component.get_description() + ', ' + self.description
```

```python
import csv
from datetime import datetime

def main():
    # Menuyu oku ve bas
    with open('Menu.txt', 'r',encoding="utf-8") as menu_file:
        menu = menu_file.read()
    print(menu)

    # Kullanıcıdan pizza ile ilgili giriş al
    pizza_type = input("Lütfen bir pizza türü seçiniz (1-4): ")
    while pizza_type not in ['1', '2', '3', '4']:
        pizza_type = input("Lütfen geçerli bir pizza türü seçiniz (1-4): ")
    pizza_type = int(pizza_type)

    sauce_type = input("Lütfen bir sos seçiniz (11-16): ")
    while sauce_type not in ['11', '12', '13', '14', '15', '16']:
        sauce_type = input("Lütfen geçerli bir sos seçiniz (11-16): ")
    sauce_type = int(sauce_type)

    # Kullanıcı girişine göre pizza tabanı seç
    if pizza_type == 1:
        pizza = KlasikPizza()
    elif pizza_type == 2:
        pizza = MargaritaPizza()
    elif pizza_type == 3:
        pizza = TurkishPizza()
    else:
        pizza = SadePizza()

    # Kullanıcı girişine göre sos ekle
    if sauce_type == 11:
        pizza = Zeytin(pizza)
    elif sauce_type == 12:
        pizza = Mantar(pizza)
    elif sauce_type == 13:
        pizza = KeciPeyniri(pizza)
    elif sauce_type == 14:
        pizza = Et(pizza)
    elif sauce_type == 15:
        pizza = Sogan(pizza)
    else:
        pizza = Misir(pizza)
#loop ve tuple kullanarak birden çok sos seçme özelliği eklenebilir 
   
    # Pizza ve sosun toplam fiyatını getir
    total_cost = pizza.get_cost()
    

    # Müşteri bilgilerini al
    name = input("İsim: ")
    tc = input("TC Kimlik Numarası: ")
    tc = int(tc)
    cc_num = input("Kredi Kartı Numarası: ")
    cc_num = int(cc_num)
    cc_cvv = input("Kredi Kartı CVV: ")
    cc_cvv = int(cc_cvv)
#int istenen değerlere str girildiğinde uyarı vermesi eklenebilir

    # Siparişi CSV dosyasına kaydet
    with open('Orders_Database.csv', mode='a', newline='') as orders_file:
        order_writer = csv.writer(orders_file)
        order_writer.writerow([name, tc, cc_num, cc_cvv, datetime.now().strftime("%Y-%m-%d %H:%M:%S"), total_cost, pizza.get_description() ])

    print("Siparişiniz alınmıştır. Toplam tutar: {} TL".format(total_cost))
    
if __name__ == '__main__':
    main()
```
